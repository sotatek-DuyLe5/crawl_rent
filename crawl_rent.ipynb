```python
import os
import json
import time
from seleniumwire import webdriver  # 💡 Thay vì selenium
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# --- Cấu hình Chrome ---
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--incognito")
options.add_argument("--window-size=1200,800")
# options.add_argument("--headless=new")  # bật nếu muốn chạy nền

seleniumwire_options = {
    'enable_har': True,
    'capture_headers': True,
    'capture_body': True,      # ✅ Bắt body response
    'verify_ssl': False
}

# Khởi tạo Selenium Wire driver
driver = webdriver.Chrome(options=options, seleniumwire_options=seleniumwire_options)
wait = WebDriverWait(driver, 20)

FB_EMAIL = os.environ.get("FB_EMAIL", "nhanmangaytho77@gmail.com")
FB_PASS = os.environ.get("FB_PASS", "xxxxxx")
GROUP_URL = "https://www.facebook.com/groups/346009719391262/?locale=vi_VN"


# ------------------- Helper -------------------
def try_find_and_click(selector, by=By.CSS_SELECTOR, timeout=8):
    try:
        el = WebDriverWait(driver, timeout).until(EC.element_to_be_clickable((by, selector)))
        el.click()
        return True
    except Exception:
        return False


# ------------------- Đăng nhập -------------------
def login_facebook(email, password):
    print("🔐 Bắt đầu đăng nhập Facebook...")
    driver.get("https://www.facebook.com/login")
    try:
        email_input = wait.until(EC.presence_of_element_located((By.ID, "email")))
        email_input.clear()
        email_input.send_keys(email)

        pass_input = wait.until(EC.presence_of_element_located((By.ID, "pass")))
        pass_input.clear()
        pass_input.send_keys(password)

        try_find_and_click("button[name='login']")

        try:
            WebDriverWait(driver, 20).until(
                EC.any_of(
                    EC.presence_of_element_located((By.CSS_SELECTOR, "a[aria-label='Trang chủ']")),
                    EC.presence_of_element_located((By.CSS_SELECTOR, "div[aria-label='Account']")),
                    EC.presence_of_element_located((By.CSS_SELECTOR, "div[role='navigation']")),
                )
            )
            print("✅ Có vẻ đã đăng nhập thành công.")
            return True
        except Exception:
            print("⚠ Không chắc đã đăng nhập — có thể bị checkpoint.")
            return False
    except Exception as e:
        print("❌ Lỗi đăng nhập:", e)
        return False


# ------------------- Bắt request GraphQL -------------------
def capture_graphql_responses(group_url, wait_after_action=5, output_file="graphql_responses.json"):
    print("🌐 Mở group:", group_url)
    driver.get(group_url)
    time.sleep(wait_after_action)

    # Cuộn xuống để kích hoạt các request GraphQL
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(wait_after_action)

    graphql_responses = []

    for request in driver.requests:
        if (
            request.response
            and "graphql" in request.url.lower()
            and "facebook.com" in request.url
        ):
            try:
                # Một số request GraphQL có header x-fb-friendly-name
                op_name = request.headers.get("x-fb-friendly-name", "UnknownQuery")
                body = request.response.body.decode("utf-8", errors="ignore")

                graphql_responses.append({
                    "url": request.url,
                    "method": request.method,
                    "status_code": request.response.status_code,
                    "operationName": op_name,
                    "headers": dict(request.headers),
                    "body": body[:5000]  # tránh file quá to
                })
            except Exception as e:
                print(f"⚠ Lỗi đọc response từ {request.url}: {e}")

    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(graphql_responses, f, ensure_ascii=False, indent=2)

    print(f"💾 Đã lưu {len(graphql_responses)} GraphQL responses vào {output_file}")

# ------------------- Chạy toàn bộ flow -------------------
try:
    ok = login_facebook(FB_EMAIL, FB_PASS)
    if not ok:
        print("⚠ Tiếp tục: script sẽ cố gắng truy cập group dù login chưa chắc thành công.")
    capture_graphql_responses(GROUP_URL, wait_after_action=6, output_file="graphql_responses.json")
finally:
    driver.quit()
```

```python
import json

# Đọc dữ liệu GraphQL responses
file_path = "graphql_responses_cdp.json"
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

def parse_multiple_jsons(s):
    """Tách nhiều JSON nối liền nhau trong 1 chuỗi"""
    decoder = json.JSONDecoder()
    idx = 0
    results = []
    while idx < len(s):
        s = s.lstrip()
        try:
            obj, end = decoder.raw_decode(s)
            results.append(obj)
            s = s[end:]  # bỏ phần đã đọc
        except json.JSONDecodeError:
            break
    return results

rows = []

for i, item in enumerate(data):
    body_raw = item.get("body")
    if not body_raw:
        continue

    jsons = parse_multiple_jsons(body_raw)
    for js in jsons:
        try:
            node = js["data"]["node"]

            # (1) Người đăng
            actors = node.get("actors", [])
            if actors and isinstance(actors, list):
                actor = actors[0]
                name = actor.get("name", "")
                url = actor.get("url", "")
                poster = f"{name} - {url}" if name or url else None
            else:
                poster = None

            # (2) Nội dung bài viết
            story_msg = (
                node.get("comet_sections", {})
                .get("content", {})
                .get("story", {})
                .get("message", {})
            )
            msg_text = None
            if isinstance(story_msg, dict):
                msg_text = story_msg.get("text") or story_msg.get("text_with_entities", {}).get("text")

            # (3) Link bài viết
            permalink = node.get("permalink_url", None)

            rows.append({
                "poster": poster,
                "message": msg_text,
                "permalink": permalink
            })

        except Exception:
            continue

print(f"✅ Trích xuất được {len(rows)} bài viết hợp lệ")

# Xem trước 3 dòng đầu
for r in rows:
    print(f"👤 Người đăng: {r['poster']}")
    print(f"💬 Bài viết: {r['message']}")
    print(f"🔗 Link: {r['permalink']}")
    print("-" * 80)
```

```python
import os
import json
import time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# ==================== CẤU HÌNH DRIVER ====================
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--incognito")
options.add_argument("--window-size=1200,800")
options.add_argument("--no-sandbox")
options.add_argument("--disable-dev-shm-usage")

# Bật performance logging để bắt GraphQL
options.set_capability("goog:loggingPrefs", {"performance": "ALL"})

driver = webdriver.Chrome(options=options)
wait = WebDriverWait(driver, 20)
driver.execute_cdp_cmd("Network.enable", {})

# ==================== CẤU HÌNH LOGIN ====================
FB_EMAIL = os.environ.get("FB_EMAIL", "nhanmangaytho77@gmail.com")
FB_PASS = os.environ.get("FB_PASS", "xxxxx")
GROUP_URL = "https://www.facebook.com/groups/346009719391262/?locale=vi_VN"

def try_find_and_click(selector, by=By.CSS_SELECTOR, timeout=8):
    try:
        el = WebDriverWait(driver, timeout).until(EC.element_to_be_clickable((by, selector)))
        el.click()
        return True
    except Exception:
        return False

# ------------------- Đăng nhập Facebook -------------------
def login_facebook(email, password):
    print("🔐 Bắt đầu đăng nhập Facebook...")
    driver.get("https://www.facebook.com/login")
    try:
        email_input = wait.until(EC.presence_of_element_located((By.ID, "email")))
        email_input.clear()
        email_input.send_keys(email)

        pass_input = wait.until(EC.presence_of_element_located((By.ID, "pass")))
        pass_input.clear()
        pass_input.send_keys(password)

        try_find_and_click("button[name='login']")

        try:
            WebDriverWait(driver, 20).until(
                EC.any_of(
                    EC.presence_of_element_located((By.CSS_SELECTOR, "a[aria-label='Trang chủ']")),
                    EC.presence_of_element_located((By.CSS_SELECTOR, "div[aria-label='Account']")),
                    EC.presence_of_element_located((By.CSS_SELECTOR, "div[role='navigation']")),
                )
            )
            print("✅ Có vẻ đã đăng nhập thành công.")
            return True
        except Exception:
            print("⚠ Không chắc đã đăng nhập — có thể bị checkpoint.")
            return False
    except Exception as e:
        print("❌ Lỗi đăng nhập:", e)
        return False

# ==================== CHẠY CHÍNH ====================
if not login_facebook(FB_EMAIL, FB_PASS):
    print("🚫 Không thể đăng nhập — dừng script.")
    driver.quit()
    exit()
# Mở group sau khi login
driver.get(GROUP_URL)
time.sleep(5)

# ==================== THU THẬP GRAPHQL ====================
scroll_times = 30
sleep_between_scrolls = 1.5
seen_request_ids = set()
graphql_responses = []

for i in range(scroll_times):
    driver.execute_script("window.scrollBy(0, window.innerHeight);")
    time.sleep(sleep_between_scrolls)

    try:
        logs = driver.get_log("performance")
    except Exception as e:
        print("⚠ Lỗi get_log('performance'):", e)
        logs = []

    for entry in logs:
        try:
            msg = json.loads(entry["message"])["message"]
        except Exception:
            continue

        if msg.get("method") == "Network.responseReceived":
            params = msg.get("params", {})
            response = params.get("response", {})
            url = response.get("url", "")
            if "graphql" in url.lower() and params.get("requestId"):
                request_id = params["requestId"]
                if request_id in seen_request_ids:
                    continue
                seen_request_ids.add(request_id)
                try:
                    body_obj = driver.execute_cdp_cmd("Network.getResponseBody", {"requestId": request_id})
                    body_text = body_obj.get("body", "")
                except Exception:
                    body_text = ""
                graphql_responses.append({
                    # "url": url,
                    # "status": response.get("status"),
                    # "mimeType": response.get("mimeType"),
                    # "operationName": response.get("headers", {}).get("x-fb-friendly-name", "UnknownQuery"),
                    "body": body_text
                })
    print(f"⤵ Scroll {i+1}/{scroll_times} — tổng GraphQL responses thu thập: {len(graphql_responses)}")

# ==================== LƯU FILE ====================
output_file = "graphql_responses_cdp.json"
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(graphql_responses, f, ensure_ascii=False, indent=2)

print(f"💾 Đã lưu {len(graphql_responses)} GraphQL responses vào {output_file}")
```
