```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.common.exceptions import NoSuchElementException
from selenium_stealth import stealth  
import time
import pandas as pd

# ===== C·∫•u h√¨nh Chrome =====
options = Options()
# options.add_argument("--headless")  # ch·∫°y ·∫©n tr√¨nh duy·ªát n·∫øu mu·ªën
options.add_argument("--disable-blink-features=AutomationControlled")
options.add_argument("--disable-infobars")
options.add_argument("--disable-extensions")
options.add_argument("--start-maximized")

service = Service(ChromeDriverManager().install())
driver = webdriver.Chrome(service=service, options=options)

# ===== Th√™m l·ªõp ng·ª•y trang Selenium =====
stealth(driver,
        languages=["en-US", "en"],
        vendor="Google Inc.",
        platform="Win32",
        webgl_vendor="Intel Inc.",
        renderer="Intel Iris OpenGL Engine",
        fix_hairline=True,
)

wait = WebDriverWait(driver, 10)

# ===== M·ªü trang web =====
url = "https://batdongsan.com.vn/cho-thue-nha-tro-phong-tro"
driver.get(url)

results = []
page = 1
max_pages = 36

while True:


    wait.until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, "#product-lists-web > div")))
    cards = driver.find_elements(By.CSS_SELECTOR, "#product-lists-web > div")
    print(f"  ‚Üí T√¨m th·∫•y {len(cards)} tin")

    for i, card in enumerate(cards, start=1):
        try:
            title = card.find_element(By.CSS_SELECTOR, "h3").text.strip()
        except:
            title = ""
        try:
            price = card.find_element(By.CSS_SELECTOR, "span.re__card-config-price").text.strip()
        except:
            price = ""
        try:
            address = card.find_element(By.CSS_SELECTOR, "div.re__card-location > span:nth-child(2)").text.strip()
        except:
            address = ""
        try:
            link = card.find_element(By.CSS_SELECTOR, "a").get_attribute("href")
        except:
            link = ""

        results.append({
            "page": page,
            "title": title,
            "price": price,
            "address": address,
            "link": link
        })

    # ==== Next Page ====
    try:
        # Ch·ªù t·∫•t c·∫£ c√°c n√∫t ph√¢n trang hi·ªÉn th·ªã
        pagination_buttons = wait.until(EC.presence_of_all_elements_located((
            By.CSS_SELECTOR,
            "div.re__srp-paging.js__srp-paging > div > a"
        )))

        # N√∫t cu·ªëi c√πng th∆∞·ªùng l√† n√∫t "Next"
        next_btn = pagination_buttons[-1]

        # Cu·ªôn t·ªõi n√∫t ƒë·ªÉ tr√°nh b·ªã che
        driver.execute_script("arguments[0].scrollIntoView({block: 'center'});", next_btn)
        time.sleep(1)

        try:
            next_btn.click()
        except Exception:
            # fallback n·∫øu b·ªã ch·∫∑n click
            driver.execute_script("arguments[0].click();", next_btn)

        print("‚û°Ô∏è ƒê√£ click n√∫t Next th·∫≠t s·ª± b·∫±ng Selenium")
        time.sleep(3)

        page += 1
        if page > max_pages:
            print("‚úÖ ƒê√£ ƒë·∫°t s·ªë trang t·ªëi ƒëa, d·ª´ng crawl.")
            break

    except Exception as e:
        print("‚ùå Kh√¥ng c√≤n trang k·∫ø ti·∫øp ho·∫∑c l·ªói khi chuy·ªÉn trang:", e)
        break
driver.quit()

df = pd.DataFrame(results)
df.to_csv("batdongsan_phongtro.csv", index=False, encoding="utf-8-sig")
print(f"‚úÖ ƒê√£ l∆∞u {len(df)} tin v√†o batdongsan_phongtro.csv")
```

```python
import os
import json
import time
from seleniumwire import webdriver  # üí° Thay v√¨ selenium
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

# --- C·∫•u h√¨nh Chrome ---
options = webdriver.ChromeOptions()
options.add_argument("--disable-gpu")
options.add_argument("--incognito")
options.add_argument("--window-size=1200,800")
# options.add_argument("--headless=new")  # b·∫≠t n·∫øu mu·ªën ch·∫°y n·ªÅn

seleniumwire_options = {
    'enable_har': True,
    'capture_headers': True,
    'capture_body': True,      # ‚úÖ B·∫Øt body response
    'verify_ssl': False
}

# Kh·ªüi t·∫°o Selenium Wire driver
driver = webdriver.Chrome(options=options, seleniumwire_options=seleniumwire_options)
wait = WebDriverWait(driver, 20)

FB_EMAIL = os.environ.get("FB_EMAIL", "nhanmangaytho77@gmail.com")
FB_PASS = os.environ.get("FB_PASS", "xxxxxx")
GROUP_URL = "https://www.facebook.com/groups/346009719391262/?locale=vi_VN"


# ------------------- Helper -------------------
def try_find_and_click(selector, by=By.CSS_SELECTOR, timeout=8):
    try:
        el = WebDriverWait(driver, timeout).until(EC.element_to_be_clickable((by, selector)))
        el.click()
        return True
    except Exception:
        return False


# ------------------- ƒêƒÉng nh·∫≠p -------------------
def login_facebook(email, password):
    print("üîê B·∫Øt ƒë·∫ßu ƒëƒÉng nh·∫≠p Facebook...")
    driver.get("https://www.facebook.com/login")
    try:
        email_input = wait.until(EC.presence_of_element_located((By.ID, "email")))
        email_input.clear()
        email_input.send_keys(email)

        pass_input = wait.until(EC.presence_of_element_located((By.ID, "pass")))
        pass_input.clear()
        pass_input.send_keys(password)

        try_find_and_click("button[name='login']")

        try:
            WebDriverWait(driver, 20).until(
                EC.any_of(
                    EC.presence_of_element_located((By.CSS_SELECTOR, "a[aria-label='Trang ch·ªß']")),
                    EC.presence_of_element_located((By.CSS_SELECTOR, "div[aria-label='Account']")),
                    EC.presence_of_element_located((By.CSS_SELECTOR, "div[role='navigation']")),
                )
            )
            print("‚úÖ C√≥ v·∫ª ƒë√£ ƒëƒÉng nh·∫≠p th√†nh c√¥ng.")
            return True
        except Exception:
            print("‚ö† Kh√¥ng ch·∫Øc ƒë√£ ƒëƒÉng nh·∫≠p ‚Äî c√≥ th·ªÉ b·ªã checkpoint.")
            return False
    except Exception as e:
        print("‚ùå L·ªói ƒëƒÉng nh·∫≠p:", e)
        return False


# ------------------- B·∫Øt request GraphQL -------------------
def capture_graphql_responses(group_url, wait_after_action=5, output_file="graphql_responses.json"):
    print("üåê M·ªü group:", group_url)
    driver.get(group_url)
    time.sleep(wait_after_action)

    # Cu·ªôn xu·ªëng ƒë·ªÉ k√≠ch ho·∫°t c√°c request GraphQL
    driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(wait_after_action)

    graphql_responses = []

    for request in driver.requests:
        if (
            request.response
            and "graphql" in request.url.lower()
            and "facebook.com" in request.url
        ):
            try:
                # M·ªôt s·ªë request GraphQL c√≥ header x-fb-friendly-name
                op_name = request.headers.get("x-fb-friendly-name", "UnknownQuery")
                body = request.response.body.decode("utf-8", errors="ignore")

                graphql_responses.append({
                    "url": request.url,
                    "method": request.method,
                    "status_code": request.response.status_code,
                    "operationName": op_name,
                    "headers": dict(request.headers),
                    "body": body[:5000]  # tr√°nh file qu√° to
                })
            except Exception as e:
                print(f"‚ö† L·ªói ƒë·ªçc response t·ª´ {request.url}: {e}")

    with open(output_file, "w", encoding="utf-8") as f:
        json.dump(graphql_responses, f, ensure_ascii=False, indent=2)

    print(f"üíæ ƒê√£ l∆∞u {len(graphql_responses)} GraphQL responses v√†o {output_file}")

# ------------------- Ch·∫°y to√†n b·ªô flow -------------------
try:
    ok = login_facebook(FB_EMAIL, FB_PASS)
    if not ok:
        print("‚ö† Ti·∫øp t·ª•c: script s·∫Ω c·ªë g·∫Øng truy c·∫≠p group d√π login ch∆∞a ch·∫Øc th√†nh c√¥ng.")
    capture_graphql_responses(GROUP_URL, wait_after_action=6, output_file="graphql_responses.json")
finally:
    driver.quit()
```

```python
import json

# ƒê·ªçc d·ªØ li·ªáu GraphQL responses
file_path = "graphql_responses_cdp.json"
with open(file_path, "r", encoding="utf-8") as f:
    data = json.load(f)

def parse_multiple_jsons(s):
    """T√°ch nhi·ªÅu JSON n·ªëi li·ªÅn nhau trong 1 chu·ªói"""
    decoder = json.JSONDecoder()
    idx = 0
    results = []
    while idx < len(s):
        s = s.lstrip()
        try:
            obj, end = decoder.raw_decode(s)
            results.append(obj)
            s = s[end:]  # b·ªè ph·∫ßn ƒë√£ ƒë·ªçc
        except json.JSONDecodeError:
            break
    return results

rows = []

for i, item in enumerate(data):
    body_raw = item.get("body")
    if not body_raw:
        continue

    jsons = parse_multiple_jsons(body_raw)
    for js in jsons:
        try:
            node = js["data"]["node"]

            # (1) Ng∆∞·ªùi ƒëƒÉng
            actors = node.get("actors", [])
            if actors and isinstance(actors, list):
                actor = actors[0]
                name = actor.get("name", "")
                url = actor.get("url", "")
                poster = f"{name} - {url}" if name or url else None
            else:
                poster = None

            # (2) N·ªôi dung b√†i vi·∫øt
            story_msg = (
                node.get("comet_sections", {})
                .get("content", {})
                .get("story", {})
                .get("message", {})
            )
            msg_text = None
            if isinstance(story_msg, dict):
                msg_text = story_msg.get("text") or story_msg.get("text_with_entities", {}).get("text")

            # (3) Link b√†i vi·∫øt
            permalink = node.get("permalink_url", None)

            rows.append({
                "poster": poster,
                "message": msg_text,
                "permalink": permalink
            })

        except Exception:
            continue

print(f"‚úÖ Tr√≠ch xu·∫•t ƒë∆∞·ª£c {len(rows)} b√†i vi·∫øt h·ª£p l·ªá")

# Xem tr∆∞·ªõc 3 d√≤ng ƒë·∫ßu
for r in rows:
    print(f"üë§ Ng∆∞·ªùi ƒëƒÉng: {r['poster']}")
    print(f"üí¨ B√†i vi·∫øt: {r['message']}")
    print(f"üîó Link: {r['permalink']}")
    print("-" * 80)
```
